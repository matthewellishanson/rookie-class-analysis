```python
import duckdb

DB_PATH = "assets/rookies.duckdb"
monitor_con = duckdb.connect(DB_PATH, read_only=True)
```

```python
monitor_con.execute("""
SELECT COUNT(*) AS total_games FROM rookie_games;
""").fetchall()
```

```python
monitor_con.execute("""
SELECT player_id, rookie_season, COUNT(*) AS games
FROM rookie_games
GROUP BY 1,2
ORDER BY rookie_season DESC, games DESC
LIMIT 10;
""").fetchdf()
```
